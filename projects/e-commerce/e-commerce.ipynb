### Добавление столбцов со временем и с выручкой

```python
data['month'] = data['date'].dt.month # Месяцы
#data['week'] = data['date'].dt.week # Неделя
data['day'] = data['date'].dt.day # День
data['revenue'] = data['quantity'] * data['price'] # выручка
```

```python
def weekend(day):
    "Функция для выходных\рабочих дней"
    if  1 <= day <= 5:
        return 'working day' # Рабочий день 
    else:
        return 'weekend' # Выходной
```

```python
def seasons(month):
    "Функция для времен года"
    if  1 <= month <= 2 or month == 12:
        return 'winter' # Рабочий день 
    elif 3 <= month <= 5:
        return 'spring' # Выходной
    elif 6 <= month <= 8:
        return 'summer' # Выходной
    else:
        return 'autumn' # Выходной
```

```python
data['weekday'] = data['day'].apply(weekend).astype('category')
data['season'] = data['month'].apply(seasons).astype('category')
data.sample(6, random_state=78)
```

### Матрица корреляции (уверен будет красиво)

```python
def weekend_1(day):
    "Функция для выходных\рабочих дней"
    if  1 <= day <= 5:
        return 0 # Рабочий день 
    else:
        return 1 # Выходной
```

```python
def seasons_1(month):
    "Функция для времен года"
    if  1 <= month <= 2 or month == 12:
        return 1 # Рабочий день 
    elif 3 <= month <= 5:
        return 2 # Выходной
    elif 6 <= month <= 8:
        return 3 # Выходной
    else:
        return 4 # Выходной
```

## Проверка статистических гиппотез

### Проверка выручки в разные сезоны (зима, весна, лето, осень) Не забыдь про метод Шидака тут

Возьмем за основной сезон - осень. Будем сравнивать выручку с остальными сезоанми. Всего сравнений будет 3. Применим сразу Метод Шидака 1 - (1 - 0.05)^(1/3)

```python
def stat_season(season,n):
    "Для стат значимости, где n - кол-во проверок, Метод шидака"
    
    p_value = stats.mannwhitneyu(clean_data[clean_data['season'] == 'autumn']['revenue'],
                                                          clean_data[clean_data['season'] == season]['revenue'],
                                                         alternative='two-sided')[1]
    print('P-value равно','{0:.3f}',p_value)
    if p_value < 1-(1-0.05)**(1/n):
        print('Между выручкой в autumn и ',season, 'есть разница')
    else:
        print('Между выручкой в autumn и ',season, 'нет разницы')
    print('Относительная разница в прибыли относительно autumn = ',
      '{0:.2f}'.format((clean_data[clean_data['season'] == 'autumn']['revenue'].mean()/
                       clean_data[clean_data['season'] == season]['revenue'].mean()-1)*100),'%')
```

```python
stat_season('winter',3)
```

```python
stat_season('spring',3)
```

```python
stat_season('summer',3)
```

Вывод:
По сезонам все разные. Выручка в различные сезоны разная
