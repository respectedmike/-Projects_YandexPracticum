## Предобработка данных

<h3> Предобработка столбца 'ceiling_height'</h3>
Приступим к предобработке высоты потолка 

```python
df['ceiling_height'].isna().sum()
```

```python
df[df['ceiling_height']<2]
#Судя по наличию фотографий и более менее приемлимой цене, здесь просто не хватает 
#1 метра добавим его.
```

```python
# Функция для добавления метра 
def plus_metr(visota):
    if visota <2:
        return visota + 1
    else:
        return visota
```

```python
df['ceiling_height'] = df['ceiling_height'].apply(plus_metr)
df[df['ceiling_height']<2] # проверяем
```

Следующим шагом избавимся от аномально больших значений.

```python
df[df['ceiling_height']>5] # ну больших значений не слишком много опять же напишем функцию
```

```python
# Функция для нормализации высоты 
def norm_visota(visota):
    if 10 > visota > 5: # Высота потолка выше 5 метров я считаю аномальной
        return visota / 2
    elif 20 > visota >= 10:
        return visota / 10 + 1
    elif 50 > visota >=20:
        return visota / 10 
    elif visota >= 50:
        return 2.5
    else:
        return visota
```

```python
df['ceiling_height'] = df['ceiling_height'].apply(norm_visota)
df[df['ceiling_height']>5] # проверяем
```

Теперь разберемся с пропусками в данном столбце. Оптимальным решением сччитаю написать функцию, которая будет высчитывать медианное значение по расположению и заменять пропуск.

```python
# Можно обьеденить все местоположения, где меньше 10 обьявление в другие
group_names = df['locality_name'].value_counts()
group_names2 = group_names[group_names>2].index
group_names2
```

```python
df['group_locality_name'] =  df['locality_name'].where(
    df['locality_name'].isin(group_names2), 'Другие'
)
```

```python
df['group_locality_name'].value_counts()# 
```

```python
med_height = df.pivot_table(
    index='group_locality_name', values='ceiling_height', aggfunc=['median']
)
med_height = med_height.reset_index()
med_height.columns = ['group_locality_name', 'med_ceiling_height']
med_height
```

```python
df = df.merge(med_height, how = 'left')
```

```python
# Функция для заполнения пустых значений в столбце 
# высота потолка исходя из медианного значения расположения
def med_height(row):
    med_height = row['med_ceiling_height']
    height = row['ceiling_height']
    if pd.isna(height) == True:
        height = med_height
    return height
```

```python
df['ceiling_height'] = df.apply(med_height, axis = 1) # применяем функцию
df.info()
```

```python
# Функция для заполнения пустых значений в столбце 
# площадь кухни 
def kitchen(row):
    kitch = row['kitchen_area']
    total = row['total_area']
    living = row['living_area']
    room = row['rooms']
    if (pd.isna(kitch) == True) and (pd.isna(living) != True):
        kitch = 0.9*total - living
    elif (pd.isna(kitch) == True) and (pd.isna(living) == True):
        kitch = 0.19*total
    return kitch
```
